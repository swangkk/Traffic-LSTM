```python
wimdata['검측일시']=wimdata['검측일시'].astype(str)
```

```python
wimdata['time'] = wimdata['검측일시'].str[0:14]
```

```python
wimdata['speed'] = wimdata['속도'] / 100
```

```python
# Year와 Make를 기준으로 (kW) 컬럼의 평균값을 구한다.
# 그 다음 피벗 테이블을 만드는 데 index(행) = YEAR, column(열) = Make 로 한다.
pivot = wimdata.pivot_table(values=['speed','검측일시'], index=['date','Hour'], columns='차로번호', aggfunc={'speed':np.mean, '검측일시':'count'})
```
